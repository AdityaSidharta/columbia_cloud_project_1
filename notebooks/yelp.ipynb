```python
import requests
import pandas as pd
```

```python
#purposefully changed API_KEY to maintain privacy
API_KEY = 'https://www.yelp.com/developers/v3/manage_app?app_created=True'
```

```python
TERMS = [
    'chinese food',
    'american food',
    'japanese food',
    'french food'
]

LOCATIONS = [
    'manhattan',
    'queens',
    'brooklyn'
]
```

```python
my_headers = {'Authorization' : 'Bearer {}'.format(API_KEY)}
```

```python
OFFSETS = range(0, 1000, 50)
```

```python
URL = "https://api.yelp.com/v3/businesses/search"
LIMIT = 50
```

```python
FINAL_OUTPUT = []
```

```python
def is_valid(raw):
    if 'id' not in raw:
        return False
    if 'name' not in raw:
        return False
    if 'location' not in raw:
        return False
    if 'coordinates' not in raw:
        return False
    if 'review_count' not in raw:
        return False
    if 'rating' not in raw:
        return False
    return True
        
def get_info(raw, query, location):
    output = {
        'id': raw['id'],
        'name': raw['name'],
        'address': ', '.join(raw['location']['display_address']),
        'coordinates': (raw['coordinates']['latitude'], raw['coordinates']['longitude']),
        'review_count': raw['review_count'],
        'rating': raw['rating'],
        'zip_code': raw['location']['zip_code'],
        'query': query,
        'location': location
    }
    return output
```

```python
for term in TERMS:
    for location in LOCATIONS:
        for offset in OFFSETS:
            print("term : {}, location : {}, offset : {}".format(term, location, offset))
            r = requests.get(URL, 
                 params={"term": term, "location": location, "limit": LIMIT, "offset": offset}, 
                 headers=my_headers)
            for business in r.json()['businesses']:
                if is_valid(business):
                    output = get_info(business, term, location)
                    FINAL_OUTPUT.append(output)
```

```python
df = pd.DataFrame(FINAL_OUTPUT)
```

```python
filter_df = df.drop_duplicates('id')
```

```python
filter_df.to_csv("yelp.csv", index=False)
```
